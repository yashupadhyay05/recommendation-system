```python
import re
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
udemy['description'] = udemy['Short Description'].apply(str)
```

```python
udemy['description'] = udemy['description'].str.lower()
```

```python
udemy['description'] = udemy['description'].apply(remove_punctuations)
```

```python
udemy['description'] = udemy['description'].str.replace('\d+', '')
```
